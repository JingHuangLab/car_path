#### b. Calculations

calculations are at ./ms0_car/fig_pyscf/car_paths/rxn1

```python
basedir = "./ms0_car/fig_pyscf/car_paths/rxn1"

def load_ener(which_path: str):
    """Load the energy vector, which_path = 'cimg' or 'mini'. """
    if not which_path in ['cimg', 'mini']:
        raise ValueError('wrong which_path spec. ')
    
    lines = open(f'{basedir}/{which_path}_ener.log', 'r').readlines()
    ener = []
    l = lines[-1].strip().split(':')[-1]
    for _ in range(7):
        ener.append(float(l.split()[_]))


    ener  = numpy.asarray(ener)
    ener -= numpy.min(ener)
    
    cvec = numpy.load(f'{basedir}/{which_path}_finl.npy')
    
    angl  = []

    for c in cvec:
        hc = c[6:]  - c[0:3]
        cn = c[0:3] - c[3:6]
        a = numpy.sum(hc*cn) / numpy.linalg.norm(hc) / numpy.linalg.norm(cn)
        if numpy.abs(a) >= 1: a /= numpy.abs(a)
        a = numpy.arccos(a)
        a /= numpy.pi
        angl.append(a)

    return numpy.asarray(ener), numpy.asarray(angl)

def load_traj():
    """Load the constraint convergence log. """
    lines = open(f'{basedir}/mini_cons.log', 'r').readlines()
    
    itr = []
    rms = []
    for l in lines:
        words = l.split()
        itr.append(int(words[1]))
        rms.append(float(words[2]))
    
    return itr, rms
```

#### c. mk plots.

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5,4), dpi=900)
plt.subplots_adjust(left=1./5., right=3.4/5., top=3.4/4., bottom=1./4.)

ener0, angl0 = load_ener('mini')
print(angl0)
fwb_ener0 = numpy.max(ener0)
bwb_ener0 = numpy.max(ener0) - ener0[0]

ax.plot(angl0, ener0, lw=1.5, color="#9A1663", zorder=5, marker='.', markersize=7, label='CAR-PT')
ax.text(0.96, 0.15, r"$\Delta E_{forward}$ = "+f"{fwb_ener0:.2f}"+"\n"+r"$\Delta E_{reverse}$ = "+f"{bwb_ener0:.2f}", 
        c="#9A1663", ha='right', va='bottom', transform=ax.transAxes, fontsize=8)


ener1, angl1 = load_ener('cimg')
print(angl1)
fwb_ener1 = numpy.max(ener1)
bwb_ener1 = numpy.max(ener1) - ener1[0]

ax.plot(angl1, ener1, lw=1.5, color="#FF97C1", zorder=5, marker='.', markersize=7, label='CI-CAR-PT')
ax.text(0.96, 0.04, r"$\Delta E_{forward}$ = "+f"{fwb_ener1:.2f}"+"\n"+r"$\Delta E_{reverse}$ = "+f"{bwb_ener1:.2f}", 
        c="#FF97C1", ha='right', va='bottom', transform=ax.transAxes, fontsize=8)

print(ener0, '\n', ener1)

ax.set( xlim=(-.05, 1.05), xticks=numpy.arange(0, 1.1, 0.5), 
        xticklabels=[r"$\pi$", r"$\frac{\pi}{2}$", r"0"], 
        ylim=(-2.5, 52.5),   yticks=[0, 25, 50])

ax.set_xlabel(r"H-C-N Angle")
ax.set_ylabel(r"Energy ($kcal\ mol^{-1}$)")
ax.tick_params(direction='in', width=1.5, length=4, labelsize=12)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.legend(loc="upper left", bbox_to_anchor=(0., 1.), frameon=False)


plt.savefig(f'{basedir}/../../plot_hcniso_b3lyp_ener.jpeg')
```
